```python
# Contoh 2.14 For loop

makanan =['nasi goreng','ayam goreng','bakso','mie ayam']
for jenis in makanan:
    print(makanan)
    
i=0
while i < len(makanan):
    print(makanan[i])
    i+=1
```

```python
# Contoh 2.14 For loop
# Menghitung jumlah elemen dalam sebuah daftar

daftar_angka = [1, 5, 8, 15, 25, 28, 30, 40, 50]
jumlah = 0
for angka in daftar_angka:
    jumlah += angka
print("Jumlah:", jumlah)
```

```python
# Contoh 2.14 For loop

kalimat = 'bunga sakura warnanya sangat cantik'
kalimat_upper = ''
kalimat_lower = ''

for karakter in kalimat:
    print(kalimat)
    
for karakter in kalimat:
    kalimat_upper += karakter.upper()
    kalimat_lower += karakter.lower()
    
print(kalimat_upper)
print(kalimat_lower)
print('bunga' in kalimat)
print('BUNGA' in kalimat)
print(kalimat)
```

```python
# Contoh 2.14 For loop
# Mengkombinasikan kedua variabel yang berbeda

perlengkapan_sekolah = ['pensil', 'pulpen', 'tas', 'buku' ]
warna = ['merah', 'putih', 'kuning', 'biru']

for jenis in perlengkapan_sekolah:
    for kategori in warna:
        print(jenis, kategori)
```

```python
# Contoh codingan gabungan dari 2.12 List di Python dan 2.13 Metoda List

angka_1 = [2, 4, 3]
angka_2 = [4, 2, 7]
angka_3 = [7, 5, 9]

# Menggabungkan beberapa variabel menjadi satu
gabungan_angka = angka_1 + angka_2 + angka_3
print("Gabungan dari angka1, angka2 dan angka3:", gabungan_angka)

print(len(gabungan_angka))
print(gabungan_angka[5])
print(gabungan_angka[5-2])
print(gabungan_angka[5*2-4])

# Menambahkan variabel baru dari angka gabungan
gabungan_angka.append(8)
print("Setelah menambahkan elemen baru:", gabungan_angka)

# Menghapus variabel tertentu dari angka gabungan
gabungan_angka.remove(7)
print("Setelah menghapus angka 7:", gabungan_angka)
```

```python
# Contoh codingan gabungan dari 2.12 List di Python, 2.13 Metoda List dan 2.14 For loop

jenis_warna = ['merah', 'putih', 'kuning', 'hijau', 'biru', 'ungu', 'pink']

# Menampilkan list jenis warna
print("List jenis warna:", jenis_warna)

# Menambahkan jenis warna baru ke dalam list 
jenis_warna.append("hitam")

# Menghilangkan jenis warna tertentu dari list
jenis_warna.remove("biru")

# Menggunakan for loop untuk mengiterasi melalui setiap jenis warna dalam list
for warna in jenis_warna:
    print("Jenis warna setelah modifikasi:", warna)

# Menghitung total jenis warna dalam list
jumlah_warna = len(jenis_warna)
print("Jumlah jenis warna dalam list:", jumlah_warna)

# Membuat list baru yang berisi panjang setiap jenis warna dalam list
panjang_warna = [len(warna) for warna in jenis_warna]
print("Panjang setiap jenis warna dalam list:", panjang_warna)
```
