```python
# Menentukan persamaan kuadrat, berdasarkan akar-akarnya yang belum diketahui:
# Langkah-langkah dalam mencari persamaan kuadratnya :
# 1. Tentukan persamaan kuadrat dari akar-akarnya
# 2. Definisikan persamaan kuadrat
# 3. Gambarkan persamaan kuadratnya


import cmath
import numpy as np
import matplotlib.pyplot as plt

# Nilai yang diberikan :
a_98 = 8
b_98 = 4
D_98 = 10

# Rumus menghitung akarnya
akar1_98 = (-b_98 - cmath.sqrt(D_98)) / (2 * a_98)
akar2_98= (-b_98 + cmath.sqrt(D_98)) / (2 * a_98)

print('Akar 1:', akar1_98)
print('Akar 2:', akar2_98)

# Memastikan akar 1 dan akar 2, bilangan real
if akar1_98.imag == 0 and akar2_98.imag == 0:
    akar1_98 = akar1_98.real
    akar2_98 = akar2_98.real
else:
    print('Akarnya sangat kompleks.')
  
# Menghitung koefisien persamaan kuadrat, y = x^2 + bx + c
koefisien_b_98 = -(akar1_98 + akar2_98)
koefisien_c_98 = akar1_98 * akar2_98

# Mendefinisikan fungsi kuadrat
def quadratic_function(x_98):
    return x_98**2 + koefisien_b_98 * x_98 + koefisien_c_98

# Menghasilkan nilai x
x_98 = np.linspace(-10, 10, 400)
# Menghitung Nilai y
y_98 = quadratic_function(x_98)

# plot untuk fungsi kuadrat
plt.plot(x_98, y_98, label=f'y = x_98^2 + ({koefisien_b_98})x_98 + ({koefisien_c_98})')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('Quadratic Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Menghitung koordinat titik puncak dari suatu fungsi kuadrat f(x)=ax^2 + bx + c
# MenentukaN posisi titik puncak ( di kiri sumbu y, pada sumbu y: dimana b=0, di kanan sumbu y)
# menghitung nilai y dari titik puncak dengan memasukkan nilai x ke dalam persamaan kuadrat

# Berikut adalah langkah-langkah untuk menentukan kondisi parabola saat memotong sumbu y:
# 1. Potong sumbu y positif: Jika c>0
# 2. Potong sumbu y di pusat (0,0): Jika c=0
# 3. Potong sumbu y negatif: Jika c<0


import numpy as np
import matplotlib.pyplot as plt

# Koefisien fungsi kuadrat
a_koefisien_98 = 1  # Koefisien x^2
b_koefisien_98 = 5  # Koefisien x
c_konstanta_98 = 3  # Konstanta


# Menentukan kondisi parabola saat memotong sumbu y
if c_konstanta_98 > 0:
    kondisi_y_98 = 'Potong sumbu y positif'
elif c_konstanta_98 == 0:
    kondisi_y_98 = 'Potong sumbu y di pusat (0,0)'
else:
    kondisi_y_98 = 'Potong sumbu y negatif'

print(f'Parabola {kondisi_y_98}.')


def posisi_titik_puncak_98(a_koefisien_98, b_koefisien_98, c_konstanta_98):
    # Menghitung x dari titik puncak
    x_puncak_98 = -b_koefisien_98 / (2 * a_koefisien_98)
    
    # Menentukan posisi titik puncak
    if x_puncak_98 < 0:
        posisi_98 = 'di kiri sumbu y'
    elif x_puncak_98 == 0:
        posisi_98 = 'tepat di sumbu y'
    else:
        posisi_98 = 'di kanan sumbu y'
    
    # Menghitung y dari titik puncak
    y_puncak_98 = a_koefisien_98 * x_puncak_98**2 + b_koefisien_98 * x_puncak_98 + c_konstanta_98
    
    return x_puncak_98, y_puncak_98, posisi_98

# Menghitung posisi titik puncak
x_puncak_98, y_puncak_98, posisi_98 = posisi_titik_puncak_98(a_koefisien_98, b_koefisien_98, c_konstanta_98)
print(f'Titik puncak berada pada koordinat ({x_puncak_98}, {y_puncak_98}) dan {posisi_98}.')

# Membuat rentang nilai x
x_1_98 = np.linspace(-10, 10, 400)

# Menghitung nilai y berdasarkan fungsi kuadrat
y_2_98 = a_koefisien_98 * x_1_98**2 + b_koefisien_98 * x_1_98 + c_konstanta_98

# Membuat grafik
plt.figure(figsize=(8, 6))
plt.plot(x_1_98, y_2_98, label=f'{a_koefisien_98}x^2 + {b_koefisien_98}x + {c_konstanta_98}')
plt.scatter(x_puncak_98, y_puncak_98, color='red')  # Menandai titik puncak dengan titik merah
plt.text(x_puncak_98, y_puncak_98, f'({x_puncak_98:.2f}, {y_puncak_98:.2f})\n{posisi_98}', 
         fontsize=12, ha='right', color='red')  # Menambahkan teks posisi
plt.title('Grafik Fungsi Kuadrat')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
# Menentuka Grafik fungsi kuadrat dan menentukan jenis akar berdasarkan diskriminan dari persamaan kuadrat
# Menghitung diskriminan dari persamaan kuadrat menggunakan rumus b**2  4*a*c
# Berdasarkan nilai diskriminan, maka menentukan jenis akar:
# 1. Jika diskriminan > 0, jenis akar adalah 'Akar real dan berbeda'
# 2. Jika diskriminan = 0, jenis akar adalah 'Akar real dan sama'
# 3. Jika diskriminan < 0, jenis akar adalah 'Akar tidak real (imajiner)'

import numpy as np
import matplotlib.pyplot as plt

# Input nilai a, b, c
a_1_98 = float(input('Masukkan nilai a: '))
b_2_98 = float(input('Masukkan nilai b: '))
c_3_98 = float(input('Masukkan nilai c: '))

def jenis_akar_98(a_1_98, b_2_98, c_3_98):
    # Menghitung diskriminan
    D_98 = b_2_98**2 - 4*a_1_98*c_3_98
    
    # Menentukan jenis akar berdasarkan diskriminan
    if D_98 > 0:
        return 'Persamaan kuadrat memiliki dua akar real yang berbeda.', D_98
    elif D_98 == 0:
        return 'Persamaan kuadrat memiliki satu akar real (akar kembar).', D_98
    else:
        return 'Persamaan kuadrat memiliki Akar tidak real (imajiner).', D_98
    
def plot_kuadrat_98(a_1_98, b_2_98, c_3_98):
    # Membuat array x untuk nilai-nilai x
    x_98 = np.linspace(-10, 10, 400)
    # Menghitung nilai y berdasarkan persamaan kuadrat
    y_98 = a_1_98 * x_98**2 + b_2_98 * x_98 + c_3_98
    
    # Plot grafik persamaan kuadrat
    plt.figure(figsize=(8, 6))
    plt.plot(x_98, y_98, label=f'{a_1_98}x^2 + {b_2_98}x + {c_3_98}')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.title('Grafik Persamaan Kuadrat')
    plt.xlabel('x')
    plt.ylabel('y')

    # Menentukan jenis akar dan diskriminan
    hasil_98, D_98 = jenis_akar_98(a_1_98, b_2_98, c_3_98)
    print(hasil_98)
    print(f'Diskriminan (D): {D_98}')

    # Menampilkan akar jika real
    if D_98 >= 0:
        akar_1_98 = (-b_2_98 + np.sqrt(D_98)) / (2*a_1_98)
        akar_2_98 = (-b_2_98 - np.sqrt(D_98)) / (2*a_1_98)
        plt.plot(akar_1_98, 0, 'ro')  # Menandai akar pertama dengan titik merah
        plt.plot(akar_2_98, 0, 'ro')  # Menandai akar kedua dengan titik merah
        plt.text(akar_1_98, 0, f'({akar_1_98:.2f}, 0)', fontsize=12, ha='right')
        plt.text(akar_2_98, 0, f'({akar_2_98:.2f}, 0)', fontsize=12, ha='left')

    plt.legend()
    plt.show()

# Plot grafik persamaan kuadrat
plot_kuadrat_98(a_1_98, b_2_98, c_3_98)
```

```python
# Menentukan titik puncak dan sumbu simetri
# Dimana f(x) = ax^2 + bx + c
# Rumus titik puncak(h,k) = -b/2a dan f(h)
# Rumus mencari sumbu simetri = (h), dimana didefinisikan sebagai garis vertikal pada koordinat x=h

import numpy as np
import matplotlib.pyplot as plt

# Koefisien fungsi kuadrat
a1_98 = 1   # Koefisien x^2
b2_98 = -3  # Koefisien x
c3_98 = 2   # Konstanta

# Fungsi kuadrat
def fungsi_kuadrat_98(x_98):
    return a1_98 * x_98**2 + b2_98 * x_98 + c3_98

# Menentukan titik puncak (h, k)
h_98 = -b2_98 / (2 * a1_98)
k_98 = fungsi_kuadrat_98(h_98)

# Menentukan sumbu simetri
sumbu_simetri_98 = h_98

# Membuat rentang nilai x
x_98 = np.linspace(-10, 10, 400)
y_98 = fungsi_kuadrat_98(x_98)

# Membuat grafik
plt.figure(figsize=(8, 6))
plt.plot(x_98, y_98, label=f'{a1_98}x^2 + {b2_98}x + {c3_98}')
plt.scatter(h_98, k_98, color='red')  # Titik puncak
plt.axvline(x=sumbu_simetri_98, color='green', linestyle='--', label=f'Sumbu Simetri x = {sumbu_simetri_98}')
plt.title('Grafik Fungsi Kuadrat')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()

# Menampilkan hasil
print(f'Titik puncak: ({h_98}, {k_98})')
print(f'Sumbu simetri: x = {sumbu_simetri_98}')
```

```python
# Menentukan persamaan kuadrat, jika dikethaui titik puncaknya(vertex)
# Harus mengetahui koordinat titik puncak (h,k) dan juga satu titik lain yang dilalui oleh parabola
# Langkah-langkahnya adalah :
# 1. Dimana persamaan bentuk vertex: y=a(xh)^2+k
# 2. Menggunakan titik lain (x1,y1) untuk mencari nilai a
# 3. Setelah a diketahui, kita dapat mengkonversi persamaan bentuk vertex ke bentuk umum y=ax^2+bx+c dengan menggunakan: 
     # 1. b = -2ah
     # 2. c = a(h^2) + k
        
        
import numpy as np
import matplotlib.pyplot as plt

# Contoh penggunaan
h_1_98 = 2  # koordinat x dari titik puncak
k_2_98 = 3  # koordinat y dari titik puncak
x1_98 = 4  # koordinat x dari titik lain
y1_98 = 11  # koordinat y dari titik lain

# Untuk menghitung koefisien a, b dan c dari persamaan kuadrat berdasarkan titik puncak (h,k) dan titik lain (x1,y1)
def find_quadratic_equation_98(h_1_98 , k_2_98, x1_98, y1_98):
    # Menghitung nilai a
    a_satu_98 = (y1_98 - k_2_98) / ((x1_98 - h_1_98) ** 2)
    
    # Menghitung nilai b dan c
    b_dua_98 = -2 * a_satu_98 * h_1_98 
    c_tiga_98 = a_satu_98 * h_1_98 **2 + k_2_98
    
    return a_satu_98, b_dua_98, c_tiga_98

# Untuk menggambar grafik persamaan kuadrat menggunakan matplotlib
def plot_quadratic(a_satu_98, b_dua_98, c_tiga_98, h_1_98, k_2_98, x1_98, y1_98):
    # Membuat rentang nilai x
    x_98 = np.linspace(h_1_98 - 10, h_1_98 + 10, 400)
    # Menghitung nilai y berdasarkan persamaan kuadrat
    y_98 = a_satu_98 * x_98**2 + b_dua_98 * x_98 + c_tiga_98
    
    # Membuat plot grafik
    plt.figure(figsize=(8, 6))
    plt.plot(x_98, y_98, label=f'y = {a_satu_98:.2f}x^2 + {b_dua_98:.2f}x + {c_tiga_98:.2f}')
    plt.scatter([h_1_98, x1_98], [k_2_98, y1_98], color='red')  # Titik puncak dan titik lain
    plt.scatter(h_1_98, k_2_98, color='blue', label=f'Titik puncak ({h_1_98}, {k_2_98})')
    plt.scatter(x1_98, y1_98, color='green', label=f'Titik ({x1_98}, {y1_98})')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.title('Grafik Persamaan Kuadrat')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

# untuk menyimpan nilai koefisien a, b, dan c dari persamaan kuadrat ke dalam variabel dengan nama yang sesuai dengan konteks tertentu
a_satu_98, b_dua_98, c_tiga_98 = find_quadratic_equation_98(h_1_98, k_2_98, x1_98, y1_98)
print(f'Persamaan kuadratnya adalah: y = {a_satu_98}x^2 + {b_dua_98}x + {c_tiga_98}')

# Menampilkan grafik
plot_quadratic(a_satu_98, b_dua_98, c_tiga_98, h_1_98, k_2_98, x1_98, y1_98)
```

```python
# Menentukan persamaan kuadrat y = ax^2 + bx + c, jika dikethaui sumbu simetrinya
# Langkah-langkah untuk menentukan persamaan kuadrat:
# 1. Gunakan sumbu simetri (misalkan x=h)
# 2. Gunakan dua titik lain (x1,y1) dan (x2,y2) yang dilalui oleh parabola untuk membuat sistem persamaan
# 3. Menyelesaikan sistem persamaan untuk mendapatkan koefisien a, b dan c


import numpy as np
import matplotlib.pyplot as plt

# Contoh penggunaan
h_satu_98 = 2  # Sumbu simetri
x1_a_98, y1_a_98 = 1, 2  # Titik pertama
x2_b_98, y2_b_98 = 3, 8  # Titik kedua

# Fungsi untuk menentukan persamaan kuadrat berdasarkan sumbu simetri dan dua titik lain
def temukan_persamaan_kuadrat_dari_sumbu_simetri(h_satu_98, x1_a_98, y1_a_98, x2_b_98, y2_b_98):
    # Sistem persamaan:
    # y1 = a*x1^2 + b*x1 + c
    # y2 = a*x2^2 + b*x2 + c
    # h = -b / (2*a) (dari sumbu simetri)

    # Membuat matriks dan vektor untuk menyelesaikan sistem persamaan
    A_98 = np.array([
        [x1_a_98**2, x1_a_98, 1],
        [x2_b_98**2, x2_b_98, 1],
        [1, -2*h_satu_98, h_satu_98**2]
    ])
    B_98 = np.array([y1_a_98, y2_b_98, 0])

    # Menyelesaikan sistem persamaan untuk mendapatkan a, b, c
    a_98, b_98, c_98 = np.linalg.solve(A_98, B_98)
    
    return a_98, b_98, c_98

# Fungsi untuk menggambar grafik persamaan kuadrat menggunakan matplotlib
def plot_quadratic(a_98, b_98, c_98, h_satu_98, x1_a_98, y1_a_98, x2_b_98, y2_b_98):
    # Membuat rentang nilai x
    x_98 = np.linspace(h_satu_98 - 10, h_satu_98 + 10, 400)
    # Menghitung nilai y berdasarkan persamaan kuadrat
    y_98 = a_98 * x_98**2 + b_98 * x_98 + c_98
    
    # Membuat plot grafik
    plt.figure(figsize=(8, 6))
    plt.plot(x_98, y_98, label=f'y = {a_98:.2f}x^2 + {b_98:.2f}x + {c_98:.2f}')
    plt.scatter([x1_a_98, x2_b_98], [y1_a_98, y2_b_98], color='red', label='Titik-titik yang dilalui')
    plt.axvline(h_satu_98, color='blue', linestyle='--', label=f'Sumbu Simetri x = {h_satu_98}')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.title('Grafik Persamaan Kuadrat')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

# Menghitung koefisien a, b, dan c
a_98, b_98, c_98 = temukan_persamaan_kuadrat_dari_sumbu_simetri(h_satu_98, x1_a_98, y1_a_98, x2_b_98, y2_b_98)
print(f'Persamaan kuadratnya adalah: y = {a_98}x^2 + {b_98}x + {c_98}')

# Menampilkan grafik
plot_quadratic(a_98, b_98, c_98, h_satu_98, x1_a_98, y1_a_98, x2_b_98, y2_b_98)
```
