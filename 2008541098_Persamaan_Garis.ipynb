```python
# Dimana mencari persamaan garis dalam bentuk y = mx + c

# Langkah awal : Menghitung kemiringan (m) dari garis menggunakan dua titik
def hitung_kemiringan_98(x1_98, y1_98, x2_98, y2_98):
    return (y2_98 - y1_98) / (x2_98 - x1_98)

# Langkah kedua : Menghitung intercept (c) dari garis menggunakan satu titik dan kemiringan
# Intercept dalam persamaan garis lurus adalah titik di mana garis tersebut memotong sumbu y
def hitung_intercept_98(x1_98, y1_98, kemiringan):
    return y1_98 - kemiringan * x1_98

# Langkah keriga : Fungsi utama untuk menghitung persamaan garis, dimana Menggabungkan kedua fungsi sebelumnya untuk mendapatkan kemiringan dan intercept, dan mengembalikan nilai-nilai tersebut.
def persamaan_garis_98(x1_98, y1_98, x2_98, y2_98):
    m_satu_98 = hitung_kemiringan_98(x1_98, y1_98, x2_98, y2_98)
    c_satu_98 = hitung_intercept_98(x1_98, y1_98, m_satu_98)
    return m_satu_98, c_satu_98

# Langkah keempat : Memasukkan koordinat dari dua titik
print('Masukkan koordinat titik pertama:')
x1_98 = float(input('x1: '))
y1_98 = float(input('y1: '))

print("Masukkan koordinat titik kedua:")
x2_98 = float(input('x2: '))
y2_98 = float(input('y2: '))

# langkah kelima : Menghitung persamaan garis
kemiringan, intercept = persamaan_garis_98(x1_98, y1_98, x2_98, y2_98)

# Langkah keenam : Menampilkan hasil
print(f'Persamaan garis yang melalui titik ({x1_98}, {y1_98}) dan ({x2_98}, {y2_98}) adalah:')
print(f'y = {kemiringan}x + {intercept}')
```

```python
# Mencari nilai gradien (kemiringan) garis yang melalui dua titik

import matplotlib.pyplot as plt
import numpy as np

# Memasukkan rumus gradien (kemiringan) garis
def hitung_gradien_98(titik_1_98, titik_2_98):
    x1_1_98, y1_1_98 = titik_1_98
    x2_2_98, y2_2_98 = titik_2_98
    m_a_98 = (y2_2_98 - y1_1_98) / (x2_2_98 - x1_1_98)
    return m_a_98

# Definisikan dua titik
titik_A_98 = (-2, 4)
titik_B_98 = (2, -6)

# Menghitung gradien
gradien_98 = hitung_gradien_98(titik_A_98, titik_B_98)

# Mencetak hasil
print(f'Gradien garis yang melalui titik {titik_A_98} dan {titik_B_98} adalah {gradien_98}')

# Menentukan nilai x dan y untuk plot
x_values_98 = [titik_A_98[0], titik_B_98[0]]
y_values_98 = [titik_A_98[1], titik_B_98[1]]

# Membuat grafik
plt.plot(x_values_98, y_values_98, label=f'Garis melalui {titik_A_98} dan {titik_B_98}')
plt.title('Grafik Garis melalui Dua Titik')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# Menentukan persamaan garis yang melalui titik (4, 5) dan bergradien 3 dan menentukan nilai c
# Rumus persamaan garis y = mx + c

import matplotlib.pyplot as plt
import numpy as np

# Diketahui nilai Gradien
m1_98 = 3

# Titik yang diberikan
x1_satu_98, y1_dua_98 = 4, 5

# Menghitung intersep y (c)
c_a_98 = y1_dua_98 - m1_98 * x1_satu_98

# Mencetak persamaan garis
print(f'Persamaan garis yang melalui titik ({x1_satu_98}, {y1_dua_98}) dan bergradien {m1_98} adalah: y = {m1_98}x + ({c_a_98})')

# Fungsi untuk persamaan garis
def garis_98(x_1_98):
    return m1_98 * x_1_98 + c_a_98

# Membuat rentang nilai x untuk plot
x_1_98 = np.linspace(-10, 10, 400)
y_2_98 = garis_98(x_1_98)

# Membuat plot
plt.figure(figsize=(8, 6))
plt.plot(x_1_98, y_2_98, label=f'y = {m1_98}x + ({c_a_98})')
plt.scatter(x1_satu_98, y1_dua_98, color='red', zorder=5, label=f'Titik ({x1_satu_98}, {y1_dua_98})')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('Grafik Persamaan Garis')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Diketahui garis lurus yang sejajar dengan garis y = 4x - 3 dan melalui titik (4, 2). Tentukan persamaan garis lurus tersebut dan buatlah grafiknya

import matplotlib.pyplot as plt
import numpy as np

# Persamaan garis awal y = 4x - 3
m_initial_98 = 4
c_initial_98 = -3

# Menyederhanakan persamaan tersebut untuk mendapatkan bentuk (y = mx + c): dimana substitusikan titik (4,2) ke dalam persamaan untuk menemukan nilai 
# 2 = 4(4)+c
# 2 = 16+
# 2 = 16+c
#  = 216
# c = 216
#  = 14

# Jadi, persamaan garis yang di cari adalah:
#  = 4  14

# Persamaan garis yang dicari y = 4x - 14
m_parallel_98 = 4
c_parallel_98 = -14

# Membuat array nilai x untuk plot
x_vals_98 = np.linspace(0, 6, 400)
# Menghitung nilai y berdasarkan persamaan garis
y_vals_initial_98 = m_initial_98 * x_vals_98 + c_initial_98
y_vals_parallel_98 = m_parallel_98 * x_vals_98 + c_parallel_98

# Membuat grafik
plt.plot(x_vals_98, y_vals_initial_98, label=f'y = {m_initial_98}x {c_initial_98:+d}')
plt.plot(x_vals_98, y_vals_parallel_98, label=f'y = {m_parallel_98}x {c_parallel_98:+d}')
plt.plot(4, 2, 'ro')  # Titik (4, 2)
plt.text(4, 2, '  (4, 2)', verticalalignment='bottom')
plt.title('Grafik Garis Sejajar')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

```python
# Membuat grafik, dimana garis yang melalui titik (1, -3) dan sejajar dengan garis y=2x+4:

import matplotlib.pyplot as plt
import numpy as np

# Titik dan gradien yang diberikan
x1_one_98, y1_two_98= 1, -3
m_one_98 = -2

# Fungsi untuk mendapatkan persamaan garis
def garis_lurus_98(x1_one_98, y1_one_98, m_one_98):
    def persamaan_garis_98(x_one_98):
        return m_one_98 * (x_one_98 - x1_one_98) + y1_two_98
    return persamaan_garis_98

# Mendapatkan persamaan garis
persamaan_98 = garis_lurus_98(x1_one_98, y1_two_98, m_one_98)

# Fungsi untuk menampilkan persamaan garis dalam bentuk standar y = mx + b
def tampilkan_persamaan_98(x1_one_98, y1_two_98, m_one_98):
    b_two_98 = y1_two_98 - m_one_98 * x1_one_98
    persamaan_str_98 = f'y = {m_one_98}x + ({ b_two_98})'
    print(f'Persamaan garis yang melalui titik ({x1_one_98}, {y1_two_98}) dan sejajar dengan y = -2x + 4 adalah {persamaan_str_98}')
    return persamaan_str_98

# Menampilkan hasil persamaan
persamaan_str_98 = tampilkan_persamaan_98(x1_one_98, y1_two_98, m_one_98)

# Menyiapkan data untuk plot
x_one_98 = np.linspace(-10, 10, 400)
y_two_98 = persamaan_98(x_one_98)

# Membuat plot
plt.figure(figsize=(8, 6))
plt.plot(x_one_98, y_two_98, label=persamaan_str_98)
plt.scatter(x1_one_98, y1_two_98, color='red', zorder=5)
plt.text(x1_one_98, y1_two_98, f'({x1_one_98}, {y1_two_98})', fontsize=12, verticalalignment='bottom', horizontalalignment='right')

# Menambahkan label dan judul
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Grafik Persamaan Garis')
plt.legend()
plt.show()
```
