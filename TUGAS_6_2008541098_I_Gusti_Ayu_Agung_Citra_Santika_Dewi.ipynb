```python
# Contoh 2.20.5 Operasi Matematika pada Numpy

import numpy as np

a_1_98 = 8
b_2_98 = 5

# Mencari sinus dari dua bilangan
sin_a_1_98 = np.sin(a_1_98)
sin_b_2_98 = np.sin(b_2_98)

# Mencari tangen dari dua bilangan
tan_a_1_98 = np.tan(a_1_98)
tan_b_2_98 = np.tan(b_2_98)

# Mencari kosinus dari dua bilangan
cos_a_1_98 = np.cos(a_1_98)
cos_b_2_98 = np.cos(b_2_98)

# Mencari akar kuadrat dari dua bilangan
sqrt_a_1_98 = np.sqrt(a_1_98)
sqrt_b_2_98 = np.sqrt(b_2_98)

print('Nilai sin dari a:', sin_a_1_98)
print('Nilai sin dari b:', sin_b_2_98)
print('Nilai tan dari a:', tan_a_1_98)
print('Nilai tan dari b:', tan_b_2_98)
print('Nilai cos dari a:', cos_a_1_98)
print('Nilai cos dari b:', cos_b_2_98)
print('Akar kuadrat dari a:', sqrt_a_1_98)
print('Akar kuadrat dari b:', sqrt_b_2_98)
```
