```python
# Contoh 2.19 Fungsi dalam Python

# Mencari nilai keliling, luas dan volume balok
panjang_balok_98 = 8
lebar_balok_98 = 5
tinggi_balok_98 = 4

def hitung_keliling_balok_98(panjang, lebar, tinggi):
    keliling_balok_98 = 4 * (panjang + lebar + tinggi)
    return keliling_balok_98

def hitung_luas_balok_98(panjang, lebar, tinggi):
    luas_balok_98 = 2 * ((panjang * lebar) + (panjang * tinggi) + (lebar * tinggi))
    return luas_balok_98

def hitung_volume_balok_98(panjang, lebar, tinggi):
    volume_balok_98 = panjang * lebar * tinggi
    return volume_balok_98

keliling_balok_98 = hitung_keliling_balok_98(panjang_balok_98, lebar_balok_98, tinggi_balok_98)
luas_balok_98 = hitung_luas_balok_98(panjang_balok_98, lebar_balok_98, tinggi_balok_98)
volume_balok_98 = hitung_volume_balok_98(panjang_balok_98, lebar_balok_98, tinggi_balok_98)

print('Keliling balok:', keliling_balok_98)
print('Luas balok:', luas_balok_98)
print('Volume balok:', volume_balok_98)
```
