```python
#Contoh Operator if di Python

nilai = float(input('Nilai anda adalah: '))
if nilai > 70:
    print('Anda mendapatkan nilai B')
    print('Selamat ya, nilai anda sudah memuaskan')
elif nilai < 70:
    print('Anda mendapatkan nilai kurang dari B')
    print('Tetap semangat ya, perlu meningkatkan prestasi akademik anda')
```

```python
#Contoh Operator if di Python
#Misalkan terdapat sebuah nilai
nilai = 70

#menentukan apakah nilai tersebut di atas 60
if nilai > 60:
    nilai = 'B' 

#menentukan apakah nilai tersebut di atas 50 (tetapi tidak lebih dari 60)   
elif nilai > 50:
    nilai = 'C'
    
#menangkap nilai-nilai yang tidak memenuhi kondisi pertama dan kedua,    
elif nilai <= 50:
    nilai = 'D'

print("Nilai Anda adalah:", nilai)
```

```python
#Contoh Operator if di Python

# Dimana pada hari senin akan dibagikan sembuah lembar hasil dari ulangan matematika. Dan beberapa siswa mendapatkan nilai sebagai berikut :
nilai = 75

# Dimana terdapat kategori nilai berdasarkan hasil nilai yang didapatkan siswa, sebagai berikut:
if nilai >= 90:
    kategori = 'A'
elif nilai >= 80:
    kategori = 'B'
elif nilai >= 70:
    kategori = 'C'
elif nilai >= 60:
    kategori = 'D'
else:
    kategori = 'E'

print("Kategori nilai:", kategori)
```

```python
#Contoh Operator if di Python

# Terdapat umur seseorang, sebagai berikut:
umur = 25

# Kita ingin memberikan status kategori usia berdasarkan umur
if umur < 18:
    status = 'Anak-anak'
elif 18 <= umur < 65:
    status = 'Dewasa'
else:
    status = 'Lansia'

print("Status usia:", status)
```

```python
#Contoh  Operator if di Python

berat = float(input("Berapakah berat badan anda? "))
kg_atau_gr = input("Berat dalam KG atau GR? ")

if kg_atau_gr.upper() == 'KG':
    berat = berat * 1000
    print("Berat anda adalah " + str(berat) + " gr")
elif kg_atau_gr.upper() == 'GR':
    berat = berat / 1000
    print(f"Berat anda adalah {berat} kg")
```

```python
# Contoh While loop di Phyton
# Dimana menggambarkan proses pertumbuhan hewan sapi seiring berjalannya waktu:

# Dimana berat awal sapi dalam kilogram
berat_awal_sapi = 100

# Pertambahan berat hewan sapi setiap bulan dalam kilogram
pertambahan_berat_per_bulan = 5

# Perkiraan berat target hewan sapi dalam kilogram
berat_target_sapi = 300

# Dimana bulan pertama
bulan = 1

# Dimana berat saat ini
berat_saat_ini = berat_awal_sapi

# Loop untuk menentukan pertumbuhan hewan sapi
while berat_saat_ini < berat_target_sapi:
    print("Bulan ke-", bulan, ": Berat sapi saat ini:", berat_saat_ini, "kg")
    berat_saat_ini += pertambahan_berat_per_bulan
    bulan += 1

print("Sapi telah mencapai berat target.")
```

```python
#Contoh penggunaan while loop dan if pada phyton
#Dimana terdapat buah mangga dalam kedaaan belum matang dan akan matang dengan hari yang akan ditentukan

kondisi_buah_mangga = "belum matang"

# Berapa hari yang dibutuhkan buah mangga untuk menjadi matang
hari_matang = 7

# IMemasukkan hari ke-1
hari = 1

# Loop untuk menentukan kematangan buah mangga
while hari <= hari_matang:
    if hari == hari_matang:
        kondisi_buah_mangga = "matang"
        print("Buah mangga telah matang setelah", hari_matang, "hari.")
    else:
        print("Hari ke-", hari, ": Buah mangga masih", kondisi_buah_mangga)
    
    hari += 1

print("Proses kematangan buah mangga selesai.")
```

```python
#Contoh penggunaan while loop dan if pada phyton

# Memasukkan dua angka
a = float(input("Masukkan angka pertama: "))
b = float(input("Masukkan angka kedua: "))

# Loop while a tidak sama dengan b
while a != b:
    if a > b:
        print("Angka pertama lebih besar dari angka kedua.")
    elif b > a:
        print("Angka kedua lebih besar dari angka pertama.")
        
    # Memasukkan kedua angka kembali untuk perbandingan selanjutnya
    a = float(input("Masukkan angka pertama: "))
    b = float(input("Masukkan angka kedua: "))

# Setelah keluar dari loop, kedua angka sama
print("Angka pertama dan angka kedua sama.")
```
