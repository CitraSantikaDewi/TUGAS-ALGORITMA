```python
# 1. Dimana persamaan garis melewati 2 titik
# 2. Dimana diketahui gradien, tetapi tidak diketahui titik yang dilewati
# 3. Megikuti persamaan garis ax+by=c dan dx+ey=f
# Menentukan bahwa persamaan garis pertama dan kedua melewati titik asal (0, 0)
# Menyederhanakan persamaan garis menjadi bentuk y = mx karena diketahui bahwa titik yang dilalui adalah titik asal


import numpy as np
import matplotlib.pyplot as plt

kejadian_98 = float(input('Kejadian yang akan dikerjakan? '))

if kejadian_98 == 1:
    print('Setiap garis melewati dua titik yang diketahui.')
    a_98 = float(input('Nilai x dari titik pertama dari garis pertama adalah'))
    b_98 = float(input('Nilai y dari titik pertama dari garis pertama adalah'))
    c_98 = float(input('Nilai x dari titik kedua dari garis pertama adalah'))
    d_98 = float(input('Nilai y dari titik kedua dari garis pertama adalah'))

    e_98 = float(input('Nilai x dari titik pertama dari garis kedua adalah'))
    f_98 = float(input('Nilai y dari titik pertama dari garis kedua adalah'))
    g_98 = float(input('Nilai x dari titik kedua dari garis kedua adalah'))
    h_98 = float(input('Nilai y dari titik kedua dari garis kedua adalah'))

    m_1_98 = (d_98 - b_98) / (c_98 - a_98)  # Gradien garis pertama
    m_2_98 = (h_98 - f_98) / (g_98 - e_98)  # Gradien garis kedua

    # Persamaan garis pertama adalah y=m_1_98(x-a)+b=m_1_98x-m_1_98a+b
    # Persamaan garis kedua adalah y=m_2_98(x-e)+f=m_2_98x-m_2_98e+f

    c_1_98 = -m_1_98 * a_98 + b_98  # Titik potong garis pertama terhadap sumbu y
    c_2_98 = -m_2_98 * e_98 + f_98  # titik potong garis kedua terhadap sumbu y garis kedua

    print('Persamaan garis pertama adalah y={:.3f}x+{:.3f}'.format(m_1_98, c_1_98))
    print('Persamaan garis kedua adalah y={:.3f}x+{:.3f}'.format(m_2_98, c_2_98))

    if m_1_98 == m_2_98:
        if c_1_98 == c_2_98:
            print('Kedua garis berhimpitan sehingga akan terdapat tak berhingga banyaknya solusi')
            x_98 = np.linspace(-5, 5, 100)
            plt.plot(x_98, m_1_98 * (x_98 - a_98) + b_98, color='red')
            plt.plot(x_98, m_2_98 * (x_98 - e_98) + f_98, color='blue')
            plt.axvline(x=0, color='grey')
            plt.axhline(y=0, color='grey')
            plt.show()
        else:
            print('Kedua garis tidak mungkin berpotongan, sehingga tidak akan terdapat solusi.')
            x_98 = np.linspace(-5, 5, 100)
            plt.plot(x_98, m_1_98 * (x_98 - a_98) + b_98, color='red')
            plt.plot(x_98, m_2_98 * (x_98 - e_98) + f_98, color='blue')
            plt.axvline(x=0, color='grey')
            plt.axhline(y=0, color='grey')
            plt.show()
    else:
        print('Akan terdapat satu titik potong')
        tp_x_98 = (-m_2_98 * e_98 + f_98 + m_1_98 * a_98 - b_98) / (m_1_98 - m_2_98)
        tp_y_98 = m_1_98 * (tp_x_98 - a_98) + b_98
        print('Titik potong kedua garis adalah ({:.3f},{:.3f})'.format(tp_x_98, tp_y_98))

        x_98 = np.linspace(tp_x_98 - 5, tp_x_98 + 5, 100)
        plt.plot(x_98, m_1_98 * (x_98 - a_98) + b_98, color='red', label='y={:.3f}(x-{:.3f})+{:.3f}'.format(m_1_98, a_98, b_98))
        plt.plot(x_98, m_2_98 * (x_98 - e_98) + f_98, color='blue', label='y={:.3f}(x-{:.3f})+{:.3f}'.format(m_2_98, e_98, f_98))
        plt.axvline(x=0, color='grey')
        plt.axhline(y=0, color='grey')
        plt.legend(loc='best')
        plt.show()

elif kejadian_98 == 2:
    print('Setiap garis diketahui gradiennya, tetapi titik yang dilewati tidak diketahui.')
    m_1_98 = float(input('Gradien dari garis pertama adalah '))
    m_2_98 = float(input('Gradien dari garis kedua adalah '))

    # Persamaan garis pertama adalah y = m_1 * x
    # Persamaan garis kedua adalah y = m_2 * x

    print('Persamaan garis pertama adalah y={:.3f}x'.format(m_1_98))
    print('Persamaan garis kedua adalah y={:.3f}x'.format(m_2_98))

    if m_1_98 == m_2_98:
        print('Kedua garis sejajar sehingga kemungkinannya adalah tidak terdapat solusi atau terdapat tak berhingga banyaknya solusi')
        x_98 = np.linspace(-5, 5, 100)
        plt.plot(x_98, m_1_98 * x_98, color='red')
        plt.plot(x_98, m_2_98 * x_98, color='blue')
        plt.axvline(x=0, color='grey')
        plt.axhline(y=0, color='grey')
        plt.show()
    else:
        print('Akan terdapat satu titik potong')
        tp_x_98 = 0
        tp_y_98 = 0
        print('Titik potong kedua garis adalah (0, 0)')

        x_98 = np.linspace(-5, 5, 100)
        plt.plot(x_98, m_1_98 * x_98, color='red', label='y={:.3f}x'.format(m_1_98))
        plt.plot(x_98, m_2_98 * x_98, color='blue', label='y={:.3f}x'.format(m_2_98))
        plt.axvline(x=0, color='grey')
        plt.axhline(y=0, color='grey')
        plt.legend(loc='best')
        plt.show()

elif kejadian_98 == 3:
    print('Kedua garis memenuhi persamaan garis dengan bentuk ax+by=c dan dx+ey=f.')
    a_98 = float(input('Koefisien dari x pada garis pertama adalah'))
    b_98 = float(input('Koefisien dari y pada garis pertama adalah'))
    c_98 = float(input('Konstanta dari garis pertama adalah'))

    d_98 = float(input('Koefisien dari x pada garis kedua adalah'))
    e_98 = float(input('Koefisien dari y pada garis kedua adalah'))
    f_98 = float(input('Konstanta dari garis kedua adalah'))

    A_98 = np.array([[a_98, b_98], [d_98, e_98]])
    B_98 = np.array([c_98, f_98])

    if np.linalg.det(A_98) == 0:
        if np.dot(np.linalg.pinv(A_98), B_98).all() == np.linalg.pinv(A_98).all():
            print('Kedua garis berhimpitan')
        else:
            print('Kedua garis sejajar')
    else:
        solution = np.linalg.solve(A_98, B_98)
        print(f'Kedua garis berpotongan di titik ({solution[0]}, {solution[1]})')

        x_98 = np.linspace(solution[0] - 5, solution[0] + 5, 100)
        y1_98 = (c_98 - a_98 * x_98) / b_98
        y2_98 = (f_98 - d_98 * x_98) / e_98
        plt.plot(x_98, y1_98, color='red', label='{:.1f}x + {:.1f}y = {:.1f}'.format(a_98, b_98, c_98))
        plt.plot(x_98, y2_98, color='blue', label='{:.1f}x + {:.1f}y = {:.1f}'.format(d_98, e_98, f_98))
        plt.scatter(solution[0], solution[1], color='black')
        plt.axvline(x=0, color='grey')
        plt.axhline(y=0, color='grey')
        plt.legend(loc='best')
        plt.show()
else:
    print('Hanya terdapat 3 kasus. Dapat memasukkan kejadian 1, 2 atau 3.')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Mendefinisikan koefisien dari dua persamaan garis
# Misalkan persamaan pertama:  2x + 4y = 8
# Misalkan persamaan kedua: 2x - 1y = 2
# Dimana akan menghasilkan grafik yang menunjukkan kedua garis serta titik perpotongan (solusi) dari sistem persamaan tersebut

A1_98, B1_98, C1_98 = 2, 4, 8  # Koefisien dari persamaan pertama: 2x + 4y = 8
A2_98, B2_98, C2_98 = 2, -1, 2  # Koefisien dari persamaan kedua: 2x - 1y = 2

# Matriks koefisien
A_98 = np.array([[A1_98, B1_98],
              [A2_98, B2_98]])

# Matriks konstanta
B_98 = np.array([C1_98, C2_98])

# Menghitung determinan matriks A
det_A_98 = np.linalg.det(A_98)

if det_A_98 != 0:
    # Menghitung solusi
    solusi_98 = np.linalg.solve(A_98, B_98)
    x_98, y_98 = solusi_98
    print(f'Solusi: x = {x_98}, y = {y_98}')
else:
    print('Tidak dapat menemukan solusi: Matriks koefisien tidak memiliki invers (determinan nol).')
    
# Membuat grafik dari kedua garis
x_vals_98 = np.linspace(-10, 10, 400)
y_vals1_98 = (C1_98 - A1_98 * x_vals_98) / B1_98
y_vals2_98 = (C2_98 - A2_98 * x_vals_98) / B2_98

plt.figure(figsize=(10, 6))
plt.plot(x_vals_98, y_vals1_98, label='2x + 4y = 8')
plt.plot(x_vals_98, y_vals2_98, label='2x - 1y = 2')
plt.scatter(x_98, y_98, color='red', zorder=5)  # Titik perpotongan
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.title('Grafik Persamaan Garis')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Perhitungan invers dari matriks koefisien dan titik potong kedua garis
# Dmana membuat grafik dari dua persamaan garis dan menunjukkan titik potongnya

# Langkah - langkah yang dilakukan :
  # 1. Menentukan persamaan garis pertama dan kedua
  # 2. Menghitung titik potong dari dua garis
  # 3. Menggambar garis dan titik potongnya

# Terdapat 2 persamaan garis :
# 1. x + y = 2
# 2. 3x + 2y = 3

import numpy as np
import matplotlib.pyplot as plt

# Koefisien dari persamaan garis pertama
a_1_98 = 1
b_2_98 = 1

# Koefisien dari persamaan garis kedua
d_4_98 = 3
e_5_98 = 2

# Konstanta dari persamaan garis pertama dan kedua
c_3_98 = 2
f_6_98 = 3

# Mengecek apakah determinan dari matriks koefisien tidak sama dengan nol
if a_1_98 * e_5_98 - b_2_98 * d_4_98 != 0:
    # Membuat matriks koefisien dan matriks konstanta
    A_1_98 = np.array([[a_1_98, b_2_98], [d_4_98, e_5_98]])
    C_2_98 = np.array([[c_3_98], [f_6_98]])
    
    # Menghitung invers dari matriks koefisien
    Ainv_98 = np.linalg.inv(A_1_98)
    print('Invers dari matriks A:')
    print(Ainv_98)
    
    # Menghitung titik potong (x, y) dengan menggunakan matriks invers
    tp_xy_98 = np.matmul(Ainv_98, C_2_98)
    print('Titik potong kedua garis adalah:')
    print(tp_xy_98)

    # Menentukan x dan y dari titik potong
    x_intercept_98, y_intercept_98 = tp_xy_98.flatten()
    
    # Membuat garis untuk grafik
    x_values_98 = np.linspace(-1, 4, 400)
    y_values_line1_98 = (c_3_98 - a_1_98 * x_values_98) / b_2_98
    y_values_line2_98 = (f_6_98 - d_4_98 * x_values_98) / e_5_98
    
    # Membuat plot
    plt.figure(figsize=(8, 6))
    
    # Menggambar garis
    plt.plot(x_values_98, y_values_line1_98, label='x + y = 2')
    plt.plot(x_values_98, y_values_line2_98, label='3x + 2y = 3')
    
    # Menandai titik potong
    plt.scatter(x_intercept_98, y_intercept_98, color='red', zorder=5)
    plt.text(x_intercept_98, y_intercept_98, f'({x_intercept_98:.2f}, {y_intercept_98:.2f})', fontsize=12, ha='right')
    
    plt.title('Grafik Persamaan Garis dan Titik Potong')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
    plt.legend()
    plt.show()
else:
    print('Matrix singular, tidak ada solusi unik atau terdapat tak berhingga banyaknya solusi.')
```

```python
# Persamaan garis yang melewati 2 titik
# Terdapat 3 kemungkinan dari kedua persamaan garis ini:
# 1. Tidak ada  perpotongan titik (kedua garis sejajar atau sama persis)
# 2. terdapat solusi (kedua garis tersebut akan saling berpotongan di satu titik) (gradien (slope) dari kedua garis tidak sama)
# 3. Tidak terdapat perpotongan garis (tidak terdapat solusi) (kedua garis tegak lurus, sehingga tidak ada titik di mana keduanya bersilangan)


import numpy as np
import matplotlib.pyplot as plt

# Garis pertama
x1_98 = float(input('Masukkan nilai x dari titik pertama garis pertama: '))
y1_98 = float(input('Masukkan nilai y dari titik pertama garis pertama: '))
x2_98 = float(input('Masukkan nilai x dari titik kedua garis pertama: '))
y2_98 = float(input('Masukkan nilai y dari titik kedua garis pertama: '))

# Garis kedua
x3_98 = float(input('Masukkan nilai x dari titik pertama garis kedua: '))
y3_98 = float(input('Masukkan nilai y dari titik pertama garis kedua: '))
x4_98 = float(input('Masukkan nilai x dari titik kedua garis kedua: '))
y4_98 = float(input('Masukkan nilai y dari titik kedua garis kedua: '))

# Menghitung gradien
m1_98 = (y2_98 - y1_98) / (x2_98 - x1_98)
m2_98 = (y4_98 - y3_98) / (x4_98 - x3_98)

if m1_98 == m2_98:
    # Kedua garis sejajar
    if (y1_98 - m1_98 * x1_98) == (y3_98 - m2_98 * x3_98):
        print('Kedua garis sejajar, tak berhingga banyaknya solusi.')
    else:
        print('Kedua garis sejajar, tidak ada solusi.')
elif np.isinf(m1_98) and np.isinf(m2_98):
    # Kedua garis tegak lurus
    if x1_98 == x3_98:
        print('Kedua garis tegak lurus, tak berhingga banyaknya solusi.')
    else:
        print('Kedua garis tegak lurus, tidak ada solusi.')
else:
    # Terdapat satu titik potong
    x_intersect_98 = (m1_98 * x1_98 - y1_98 - m2_98 * x3_98 + y3_98) / (m1_98 - m2_98)
    y_intersect_98 = m1_98 * (x_intersect_98 - x1_98) + y1_98
    print('Titik potong kedua garis adalah ({}, {})'.format(x_intersect_98, y_intersect_98))
    plt.plot([x1_98, x2_98], [y1_98, y2_98], label='Garis pertama')
    plt.plot([x3_98, x4_98], [y3_98, y4_98], label='Garis kedua')
    plt.scatter(x_intersect_98, y_intersect_98, color='red', label='Titik Potong')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Perpotongan Garis')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# 1. Garis melewati 2 titik
# 2. Gradien dari salah satu garis diketahui, sementara gradien garis yang lain tidak diketahui.
# 3. Megikuti persamaan garis ax+by=c dan dx+ey=f

# Akan dicari titik potong dari dua garis yg memenuhi salah satu dari ketiga kasus di atas


import numpy as np
import matplotlib.pyplot as plt

kasus_98 = float(input('Kasus berapa yang akan dibuat? '))

if kasus_98 == 1:
    print('Masing-masing garis akan melewati dua titik yang diketahui.')
    a_98 = float(input('Nilai x dari titik pertama dari garis pertama adalah '))
    b_98 = float(input('Nilai y dari titik pertama dari garis pertama adalah '))
    c_98 = float(input('Nilai x dari titik kedua dari garis pertama adalah '))
    d_98 = float(input('Nilai y dari titik kedua dari garis pertama adalah '))

    e_98 = float(input('Nilai x dari titik pertama dari garis kedua adalah '))
    f_98 = float(input('Nilai y dari titik pertama dari garis kedua adalah '))
    g_98 = float(input('Nilai x dari titik kedua dari garis kedua adalah '))
    h_98 = float(input('Nilai y dari titik kedua dari garis kedua adalah '))

    m_1_98 = (d_98 - b_98) / (c_98 - a_98)  # Gradien garis pertama
    m_2_98 = (h_98 - f_98) / (g_98 - e_98)  # Gradien garis kedua

    # Persamaan garis pertama adalah y=m_1(x-a)+b=m_1x-m_1a+b
    # Persamaan garis kedua adalah y=m_2(x-e)+f=m_2x-m_2e+f

    c_1_98 = -m_1_98 * a_98 + b_98  # Titik potong garis pertama terhadap sumbu y
    c_2_98 = -m_2_98 * e_98 + f_98  # titik potong garis kedua terhadap sumbu y garis kedua

    print('Persamaan garis pertama adalah y={:.3f}x+{:.3f}'.format(m_1_98, c_1_98))
    print('Persamaan garis kedua adalah y={:.3f}x+{:.3f}'.format(m_2_98, c_2_98))

    if m_1_98 == m_2_98:
        # print('Kedua garis sejajar sehingga kemungkinannya adalah tidak terdapat solusi atau terdapat tak berhingga banyaknya solusi')
        if c_1_98 == c_2_98:
            print('Kedua garis berhimpitan sehingga akan terdapat tak berhingga banyaknya solusi')
            x_98 = np.linspace(-5, 5, 100)
            plt.plot(x_98, m_1_98 * (x_98 - a_98) + b_98, color='red')
            plt.plot(x_98, m_2_98 * (x_98 - e_98) + f_98, color='blue')
            plt.axvline(x=0, color='grey')
            plt.axhline(y=0, color='grey')
            plt.show()
        else:
            print('Kedua garis tidak mungkin berpotongan, sehingga tidak akan terdapat solusi.')
            x_98 = np.linspace(-5, 5, 100)
            plt.plot(x_98, m_1_98 * (x_98 - a_98) + b_98, color='red')
            plt.plot(x_98, m_2_98 * (x_98 - e_98) + f_98, color='blue')
            plt.axvline(x=0, color='grey')
            plt.axhline(y=0, color='grey')
            plt.show()
    else:
        print('Akan terdapat satu titik potong')
        tp_x_98 = (-m_2_98 * e_98 + f_98 + m_1_98 * a_98 - b_98) / (m_1_98 - m_2_98)
        tp_y_98 = m_1_98 * (tp_x_98 - a_98) + b_98
        print('Titik potong kedua garis adalah ({:.3f},{:.3f})'.format(tp_x_98, tp_y_98))

        x_98 = np.linspace(tp_x_98 - 5, tp_x_98 + 5, 100)
        plt.plot(x_98, m_1_98 * (x_98 - a_98) + b_98, color='red',
                 label='y={:.3f}(x-{:.3f})+{:.3f}'.format(m_1_98, a_98, b_98))
        plt.plot(x_98, m_2_98 * (x_98 - e_98) + f_98, color='blue',
                 label='y={:.3f}(x-{:.3f})+{:.3f}'.format(m_2_98, e_98, f_98))
        plt.axvline(x=0, color='grey')
        plt.axhline(y=0, color='grey')
        plt.legend(loc='best')
        plt.show()

elif kasus_98 == 2:
    print('Gradien dari salah satu garis diketahui, sementara gradien garis yang lain tidak diketahui.')
    m_1_98 = float(input('Gradien dari garis pertama adalah '))
    a_98 = float(input('Nilai x dari titik yang dilewati oleh garis pertama adalah '))
    b_98 = float(input('Nilai y dari titik yang dilewati oleh garis pertama adalah '))
    print('Persamaan garis pertama adalah y={}x+{}'.format(m_1_98, b_98 - m_1_98 * a_98))

    m_2_98 = float(input('Gradien dari garis kedua adalah '))
    print('Persamaan garis kedua adalah y={}x+c'.format(m_2_98))
    # Pada titik potong, nilai y dari kedua garis akan sama, sehingga kita punya:
    # m_1*x + b = m_2*x + c
    # x = (c - b) / (m_1 - m_2)
    # Dan nilai y pada titik potong adalah m_1*x + b atau m_2*x + c
    x_98 = (b_98 - m_2_98 * a_98) / (m_1_98 - m_2_98)
    y_98 = m_1_98 * x_98 + b_98
    print('Titik potong kedua garis adalah ({}, {})'.format(x_98, y_98))

    # Plotting
    x_vals_98 = np.linspace(x_98 - 10, x_98 + 10, 100)
    y_vals_1_98 = m_1_98 * x_vals_98 + b_98
    y_vals_2_98 = m_2_98 * x_vals_98 + (y_98 - m_2_98 * x_98)
    plt.plot(x_vals_98, y_vals_1_98, '-r', label='y={}x+{}'.format(m_1_98, b_98))
    plt.plot(x_vals_98, y_vals_2_98, '-b', label='y={}x+{}'.format(m_2_98, (y_98 - m_2_98 * x_98)))
    plt.plot(x_98, y_98, 'go')  # Titik potong
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Plot Garis')
    plt.legend()
    plt.grid(True)
    plt.show()


elif kasus_98 == 3:
    print('Kedua garis memenuhi persamaan garis dengan bentuk ax+by=c dan dx+ey=f.')
    a_98 = float(input('Koefisien dari x pada garis pertama adalah'))
    b_98 = float(input('Koefisien dari y pada garis pertama adalah'))
    c_98 = float(input('Konstanta dari garis pertama adalah'))

    d_98 = float(input('Koefisien dari x pada garis kedua adalah'))
    e_98 = float(input('Koefisien dari y pada garis kedua adalah'))
    f_98 = float(input('Konstanta dari garis kedua adalah'))
    
    if a_98/e_98 != b_98/d_98:  # Jika kedua garis bukanlah garis sejajar
        x_98 = (c_98*e_98 - b_98*f_98) / (a_98*e - b_98*d_98)
        y_98 = (a_98*f_98 - c_98*d_98) / (a_98*e_98 - b_98*d_98)
        print('Titik potong kedua garis adalah ({:.3f}, {:.3f})'.format(x_98, y_98))
        
        # Plotting
        x_vals_98 = np.linspace(x_98 - 10, x_98 + 10, 100)
        y_vals_1_98 = (c_98 - a_98 * x_vals_98) / b_98
        y_vals_2_98 = (f_98 - d_98 * x_vals_98) / e_98
        plt.plot(x_vals_98, y_vals_1_98, '-r', label='{}x+{}y={}'.format(a_98, b_98, c_98))
        plt.plot(x_vals_98, y_vals_2_98, '-b', label='{}x+{}y={}'.format(d_98, e_98, f_98))
        plt.plot(x_98, y_98, 'go')  # Titik potong
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title('Plot Garis')
        plt.legend()
        plt.grid(True)
        plt.show()
        
    else:
        if b_98*f_98 == c_98*e_98:  # Jika kedua garis identik (berimpitan)
            print('Kedua garis berimpitan.')
        else:  # Jika kedua garis paralel tapi tidak identik
            print('Kedua garis paralel dan tidak berimpitan.')

else:
    print('Hanya terdapat 3 kasus. Mohon masukkan 1, 2 atau 3 saja.')
```
