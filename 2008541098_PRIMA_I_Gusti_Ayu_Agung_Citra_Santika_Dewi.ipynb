```python
# Nim : 2008541098
# Dimana diketahui bahwa 𝑛 adalah 98, maka tentukan biangan prima antara 1000𝑛 dan 1000(𝑛+1)


# Diketahui n adalah 98
n_98 = 98

# Rentang bilangan prima yang diperiksa adalah 1000n sampai 1000(n+1)
batas_bawah_bilangan_98 = 1000 * n_98
batas_atas_bilangan_98 = 1000 * (n_98 + 1)

# Fungsi untuk memeriksa apakah sebuah bilangan adalah bilangan prima
def memeriksa_prima_98(bilangan_98):
    if bilangan_98 <= 1:  # memeriksa jika bilangan kurang dari atau sama dengan 1, yang bukan bilangan prima
        return False
    for i in range(2, int(bilangan_98 ** 0.5) + 1):  # memeriksa pembagi bilangan dari 2 hingga akar kuadrat dari bilangan tersebut
        if bilangan_98 % i == 0:  # memeriksa jika bilangan habis dibagi oleh i. Jika iya, maka bilangan tersebut bukan bilangan prima.
            return False  # Dimana jika bilangan tersebut bukan bilangan prima
    return True  # Dimana jika bilangan tersebut adalah bilangan prima

# Mencari bilangan prima dalam rentang yang ditentukan
prima_98 = []
for bilangan_98 in range(batas_bawah_bilangan_98, batas_atas_bilangan_98):
    if memeriksa_prima_98(bilangan_98):
        prima_98.append(bilangan_98)

print('Bilangan prima yang ada pada rentang n = 98, yaitu antara', batas_bawah_bilangan_98, 'dan', batas_atas_bilangan_98, 'adalah:')
print(prima_98)
```
