```python
#Contoh Strings Methods

tanaman_hias = 'Bunga Tulip'
tumbuhan = 'Pohon Beringin'

print (len(tanaman_hias))
print (len(tumbuhan))


#Mengaplikasikan metode lower() ke setiap string secara terpisah
tanaman_hias_lower = tanaman_hias.lower()
tumbuhan_lower = tumbuhan.lower()

print(tanaman_hias_lower)
print(tumbuhan_lower)


#Mengaplikasikan metode upper() ke setiap string secara terpisah
tanaman_hias_upper = tanaman_hias.upper()
tumbuhan_upper = tumbuhan.upper()

print(tanaman_hias_upper)
print(tumbuhan_upper)


#Mengaplikasikan metode find ke setiap string secara terpisah
tanaman_hias_find = tanaman_hias.find('Tulip')
tumbuhan_find = tanaman_hias.find('Beringin')

print(tanaman_hias_find)
print(tumbuhan_find)


#Mengaplikasikan metode replace ke setiap string secara terpisah

#Menggunakan penulisan yang sama
tanaman_hias_replace = tanaman_hias.replace('Bunga','Tanaman')
tumbuhan_replace = tumbuhan.replace('Pohon','Tumbuhan')

print(tanaman_hias_replace)
print(tumbuhan_replace)

#Mengaplikasikan metode replace ke setiap string secara terpisah

#Menggunakan penulisan yang berbeda
tanaman_hias_replace = tanaman_hias.replace('bunga','Tanaman')
tumbuhan_replace = tumbuhan.replace('pohon','Tumbuhan')

print(tanaman_hias_replace)
print(tumbuhan_replace)


print('Bunga' in tanaman_hias)
print('bunga' in tanaman_hias)

print('Beringin' in tumbuhan)
print('beringin' in tumbuhan)
```

```python
#Contoh Operator Perbandingan di Python
#Dimana seseorang memiliki selotip dengan jumlah yang berbeda
#Misalkan Andi memiliki 20 buah selotip dan Beni memiliki 10 selotip
#Maka dicari perbandingan dalam 2 orang yang memiliki jumlah selotip yang berbeda

#Misalkan : Andi = 'a', Beni = 'b'
a = 20
b = 10

nilai_pertama = a<b
nilai_kedua = b<=a
nilai_ketiga = a!=b
nilai_keempat = b==a

print(nilai_pertama)
print(nilai_kedua)
print(nilai_ketiga)
print(nilai_keempat)
```

```python
#Contoh Operator Perbandingan di Python

a = 15
b = 10
c = 7

nilai_pertama = a<b<c
nilai_kedua = b<=a , a<b
nilai_ketiga = a!=b>c
nilai_keempat = a==b>c
nilai_kelima = c<=a>b

print(nilai_pertama)
print(nilai_kedua)
print(nilai_ketiga)
print(nilai_keempat)
print(nilai_kelima)
```

```python
#Contoh Operator Logika di Python
#Dimana Operator and akan bernilai True jika kedua nilai adalah True
#Operator or akan bernilai True jika salah satu nilainya adalah True
#Operator not akan menghasilkan nilai berlawanan. Jika nilai variabel x adalah True, maka nilai dari not(x) adalah False.

pensil = 8
pulpen = 12

print(pensil < 7 or pulpen < 10)
print(pensil > 7 and pulpen < 10)
print(pensil < 7 or pulpen > 10)
print(not(pensil < 7 or pulpen > 12))
```
